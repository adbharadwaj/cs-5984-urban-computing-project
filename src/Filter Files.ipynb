```python
import pandas as pd
```

```python
import os
for filename in os.listdir(os.path.join('Regression')):
    if '.tsv' in filename and '8' in filename:
        df = pd.read_csv(os.path.join('Regression', filename), sep='\t')
        df = df[[
                'timestamp',
                'lat',
                'long',
                'police_factor', 
                'yelp_factor', 
                'prev_day_crime_freq',
                'prev_7_days_crime_freq',
                'pred_label',
                'label'
            ]]
        df.to_csv(os.path.join('Output', 'full_'+filename.replace('.tsv', '.csv')), index=False)
#         df = df[df.timestamp == '2006-12-08']
        df.to_csv(os.path.join('Output', 'date_'+filename.replace('.tsv', '.csv')), index=False)
```
